```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    words = text.split()

    words = [word for word in words if word not in stop_words]

    words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['clean_text'] = df['text'].apply(preprocess_text)
```

```python
df['clean_text'] = df['clean_text'].fillna("")
df = df[df['clean_text'].str.strip() != ""]
```

```python
GLOVE_PATH = "/content/glove.6B.100d.txt"
EMBED_DIM = 100

glove_dict = {}

with open(GLOVE_PATH, encoding="utf-8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], dtype="float32")
        glove_dict[word] = vector

print("Loaded GloVe words:", len(glove_dict))
```

```python
from collections import Counter

train_tokens = train_df['clean_text'].str.split()

vocab = Counter()
for tokens in train_tokens:
    vocab.update(tokens)

unique_tokens = set(vocab.keys())
print("Unique train tokens:", len(unique_tokens))
```

```python
covered_tokens = [w for w in unique_tokens if w in glove_dict]
coverage = len(covered_tokens) / len(unique_tokens) * 100

print(f"GloVe coverage: {coverage:.2f}%")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=20000)
tfidf.fit(train_df['clean_text'])

idf_dict = dict(zip(tfidf.get_feature_names_out(), tfidf.idf_))
```

```python
def document_embedding(text):
    tokens = text.split()

    doc_vec = np.zeros(EMBED_DIM)
    weight_sum = 0.0

    for word in tokens:
        if word in glove_dict and word in idf_dict:
            weight = idf_dict[word]
            doc_vec += glove_dict[word] * weight
            weight_sum += weight

    if weight_sum != 0:
        doc_vec /= weight_sum

    return doc_vec
```

```python
X_train = np.vstack(train_df['clean_text'].apply(document_embedding))
X_val   = np.vstack(val_df['clean_text'].apply(document_embedding))
X_test  = np.vstack(test_df['clean_text'].apply(document_embedding))

print("Train:", X_train.shape)
print("Val:", X_val.shape)
print("Test:", X_test.shape)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def build_regression_model(input_dim):
    model = models.Sequential([
        layers.Dense(128, activation='relu', input_shape=(input_dim,)),
        layers.Dropout(0.3),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)  # regression output
    ])

    model.compile(
        optimizer='adam',
        loss='mse',
        metrics=[tf.keras.metrics.RootMeanSquaredError()]
    )

    return model
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

def global_mean_baseline(y_train, y_test):
    mean_value = np.mean(y_train)
    preds = np.full_like(y_test, mean_value)

    mse = mean_squared_error(y_test, preds)
    rmse = np.sqrt(mse)

    return mse, rmse, mean_value
```

```python
from sklearn.metrics import mean_squared_error

def run_experiment(text_column):
    print(f"\n===== Using text column: {text_column} =====")

    X_train_col = build_embeddings(train_df, text_column)
    X_val_col   = build_embeddings(val_df, text_column)
    X_test_col  = build_embeddings(test_df, text_column)

    model = build_regression_model(EMBED_DIM)

    model.fit(
        X_train_col, y_train,
        validation_data=(X_val_col, y_val),
        epochs=20,
        batch_size=32,
        verbose=1
    )

    preds = model.predict(X_test_col).flatten()

    mse = mean_squared_error(y_test, preds)
    rmse = np.sqrt(mse)

    print(f"Test MSE  : {mse:.4f}")
    print(f"Test RMSE : {rmse:.4f}")

    return mse, rmse
```

```python
mse_overview, rmse_overview = run_experiment('overview')
```

```python
mse_keywords, rmse_keywords = run_experiment('keywords')
```

```python
mse_tagline, rmse_tagline = run_experiment('tagline')
```

```python
baseline_mse, baseline_rmse, mean_rating = global_mean_baseline(y_train, y_test)

print("\n===== Baseline (Global Mean) =====")
print("Mean rating:", round(mean_rating, 3))
print("Baseline MSE :", round(baseline_mse, 4))
print("Baseline RMSE:", round(baseline_rmse, 4))
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import ast

def parse_genres(g):
    if pd.isna(g):
        return []
    if isinstance(g, str):
        # Split the string by space and filter out empty strings
        return [genre.strip() for genre in g.split(' ') if genre.strip()]
    return []

train_genres = train_df['genres'].apply(parse_genres)
val_genres   = val_df['genres'].apply(parse_genres)
test_genres  = test_df['genres'].apply(parse_genres)

mlb = MultiLabelBinarizer()
y_train = mlb.fit_transform(train_genres)
y_val   = mlb.transform(val_genres)
y_test  = mlb.transform(test_genres)

num_labels = len(mlb.classes_)
print("Number of genres:", num_labels)
```

```python
def build_embeddings(df, text_column):
    texts = df[text_column].fillna("").apply(preprocess_text)
    return np.vstack(texts.apply(document_embedding))
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def build_multilabel_model(input_dim, num_labels):
    model = models.Sequential([
        layers.Dense(256, activation='relu', input_shape=(input_dim,)),
        layers.Dropout(0.3),
        layers.Dense(128, activation='relu'),
        layers.Dense(num_labels, activation='sigmoid')
    ])

    model.compile(
        optimizer='adam',
        loss='binary_crossentropy'
    )

    return model
```

```python
from sklearn.metrics import f1_score, hamming_loss, jaccard_score

def evaluate_multilabel(y_true, y_pred_probs, threshold=0.5):
    y_pred = (y_pred_probs >= threshold).astype(int)

    micro_f1 = f1_score(y_true, y_pred, average='micro')
    macro_f1 = f1_score(y_true, y_pred, average='macro')
    hamming  = hamming_loss(y_true, y_pred)
    jaccard  = jaccard_score(y_true, y_pred, average='samples')

    return micro_f1, macro_f1, hamming, jaccard
```

```python
def run_multilabel_experiment(text_column):
    print(f"\n===== Using text column: {text_column} =====")

    X_train_col = build_embeddings(train_df, text_column)
    X_val_col   = build_embeddings(val_df, text_column)
    X_test_col  = build_embeddings(test_df, text_column)

    model = build_multilabel_model(EMBED_DIM, num_labels)

    model.fit(
        X_train_col, y_train,
        validation_data=(X_val_col, y_val),
        epochs=15,
        batch_size=32,
        verbose=1
    )

    y_pred_probs = model.predict(X_test_col)

    micro_f1, macro_f1, hamming, jaccard = evaluate_multilabel(y_test, y_pred_probs)

    print(f"Micro-F1   : {micro_f1:.4f}")
    print(f"Macro-F1   : {macro_f1:.4f}")
    print(f"Hamming    : {hamming:.4f}")
    print(f"Jaccard    : {jaccard:.4f}")

    return micro_f1, macro_f1, hamming, jaccard
```

```python
micro_o, macro_o, ham_o, jac_o = run_multilabel_experiment('overview')
```

```python
micro_k, macro_k, ham_k, jac_k = run_multilabel_experiment('keywords')
```

```python
micro_t, macro_t, ham_t, jac_t = run_multilabel_experiment('tagline')
```

```python
import ast

def parse_genres(g):
    if pd.isna(g):
        return []
    if isinstance(g, str):
        try:
            g = ast.literal_eval(g)
            return [i['name'] for i in g]
        except:
            return []
    return []

df['genre_list'] = df['genres'].apply(parse_genres)
```

```python
genre_df = df[['clean_text', 'genre_list']].explode('genre_list')
genre_df = genre_df.dropna(subset=['genre_list'])
```

```python
genre_df['tokens'] = genre_df['clean_text'].str.split()
```

```python
from collections import Counter

genre_word_counts = {}

for genre in genre_df['genre_list'].unique():
    tokens = genre_df[genre_df['genre_list'] == genre]['tokens']

    counter = Counter()
    for token_list in tokens:
        counter.update(token_list)

    genre_word_counts[genre] = counter
```

```python
top_words_per_genre = {}

for genre, counter in genre_word_counts.items():
    top_words = counter.most_common(10)
    top_words_per_genre[genre] = top_words
```

```python
bottom_words_per_genre = {}

for genre, counter in genre_word_counts.items():
    filtered = {w: c for w, c in counter.items() if c >= 3}

    bottom_words = sorted(filtered.items(), key=lambda x: x[1])[:10]
    bottom_words_per_genre[genre] = bottom_words
```

```python
top_tables = []

for genre, words in top_words_per_genre.items():
    for word, freq in words:
        top_tables.append([genre, word, freq])

top_df = pd.DataFrame(top_tables, columns=['Genre', 'Word', 'Frequency'])
top_df.sort_values(['Genre', 'Frequency'], ascending=[True, False]).head(20)
```

```python
bottom_tables = []

for genre, words in bottom_words_per_genre.items():
    for word, freq in words:
        bottom_tables.append([genre, word, freq])

bottom_df = pd.DataFrame(bottom_tables, columns=['Genre', 'Word', 'Frequency'])
bottom_df.sort_values(['Genre', 'Frequency'], ascending=[True, True]).head(20)
```
